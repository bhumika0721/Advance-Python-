# what was the best month for sales ? how much was earned that month?

```python
result = a.groupby("Month").sum()

months=range(1,13)
plt.bar(months,result["Sales"])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

# Which city generates the most revenue?

```python
city_sales = a.groupby('City').sum()
sns.barplot(x=city_sales.index, y=city_sales['Sales'], palette="plasma")
plt.title('Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.show()
```

 # At what hours are customers most active?


```python
hourly_orders = a.groupby('Hour').count()

sns.lineplot(x=hourly_orders.index, y=hourly_orders['Order ID'], marker='o')
plt.title('Number of Orders by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()
```

# What are the most sold products by quantity?

```python
product_quantity = a.groupby('Product').sum().sort_values('Quantity Ordered', ascending=False)

sns.barplot(x=product_quantity['Quantity Ordered'], y=product_quantity.index, palette="coolwarm")
plt.title('Top-Selling Products (By Quantity)')
plt.xlabel('Quantity Ordered')
plt.ylabel('Product')
plt.show()
```

# What is the relationship between product price and quantity sold?

```python

sns.scatterplot(x='Price Each', y='Quantity Ordered', data=a)
plt.title('Price vs Quantity Ordered')
plt.xlabel('Price ($)')
plt.ylabel('Quantity Ordered')
plt.grid(True)
plt.show()
```

# Which city shows the highest variability in sales?

```python
sns.boxplot(x='City', y='Sales', data=a, palette='viridis')
plt.title('Sales Distribution by City (Box Plot)')
plt.xlabel('City')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```

# Is there a relationship between the price of a product and the quantity ordered?

```python
sns.lmplot(x='Price Each', y='Quantity Ordered', data=a, height=6, aspect=1.5, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title('Price vs Quantity Ordered (Linear Regression)')
plt.xlabel('Price ($)')
plt.ylabel('Quantity Ordered')
plt.grid(True)
plt.show()
```

# What percentage of total sales comes from each city?

```python
city_sales = a.groupby('City').sum()

plt.pie(city_sales['Sales'], labels=city_sales.index, colors=sns.color_palette('pastel'))
plt.title('Sales Contribution by City')
plt.show()
```

# What is the distribution of order quantity?




```python
sns.distplot(a['Quantity Ordered'], bins=30, color='green')
plt.title('Distribution of Quantity Ordered')
plt.xlabel('Quantity')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

# What is the distribution of sales amount?

```python
sns.histplot(a['Sales'], bins=50, kde=True, color='purple')
plt.title('Sales Amount Distribution')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

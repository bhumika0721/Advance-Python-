```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,15,7,20,25]


plt.plot(x,y, marker="o",linestyle="dashed",color="b",label="priya")
plt.xlabel("subcategory")
plt.ylabel("profit")
plt.title("subcategory wise sales")
plt.grid(True)
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,20,25,15]

plt.bar(categories , values ,color = 'g')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("bar chart example")
plt.show()
```
